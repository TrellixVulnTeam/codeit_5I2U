```python
import pandas as pd

df = pd.read_csv('data/enrolment_1.csv')
```

```python
#1,2,3번조건
condition_1 = (df['course name'] == 'information technology') & (df['year']==1)
condition_2 = (df['course name'] == 'commerce') & (df['year']==4)
condition_3 = True
```

```python
#1,2번조건에 해당되는것 not allowed 설정
df['status'] = 'allowed'
filter1 = condition_1 | condition_2
df.loc[filter1,'status'] = "not allowed"
df
```

```python
#
condition_3 = df.loc[df['status'] == "allowed","course name"].value_counts() <= 5
```

```python
condition_3
```

```python
for i in range(len(condition_3.index)):
    if condition_3[i]:
        df.loc[df['course name'] == condition_3.index[i],'status'] = 'not allowed'
```
