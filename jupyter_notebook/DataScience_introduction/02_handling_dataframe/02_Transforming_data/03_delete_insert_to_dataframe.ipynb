# 03.DataFrame에 값 추가/삭제

```python
import pandas as pd
```

```python
iphone_df = pd.read_csv('data/iphone.csv', index_col=0)
iphone_df
```

열을 한번 바꿔보자

```python
iphone_df.loc['iPhone X'] = ['2018-10-26', 6.1, '3GB','IOS 12.0.1','Yes']
iphone_df
```

그런데 열에는 없는 iphon xr 에 값을 넣으려 시도하면 어떻게 될까?

```python
iphone_df.loc['iPhone XR'] = ['2018-10-26', 6.1,'3GB','ios 12.0.1', 'Yes']
iphone_df
```

그렇다면 없던 컬럼 제조사에다가 apple 입력하면 어케 될까?

```python
iphone_df['제조사'] = 'Apple'
```

```python
iphone_df
```

삭제하는 방법을 배워보겠다

row iPhone XR 를 삭제해보자

```python
iphone_df.drop('iPhone XR',axis='index',inplace = False)
```

없애고싶은 로우 인덱스값,그리고 컬럼말고 인덱스를 선택한다는 어트리뷰트, 이건 설명안함

```python
iphone_df
```

엥 칼럼 표시해보니 안없어지고 그대로임 그이유는 아까 설명안했던 inplace = False 때문인데   
그의미는 원본 데이터프레임값을 건드리지 않겠다는 뜻이다.
만약 자체를 바꾸고싶다면 inplace = True 로 저 모듈을 실행하면 된다

```python
iphone_df.drop('iPhone XR',axis='index',inplace = True)
iphone_df
```

이번엔 제조사 칼럼을 없애보자

```python
iphone_df.drop('제조사',axis='columns',inplace = True)
iphone_df
```

이번엔 여러줄을 한번에 없애줄것임

```python
iphone_df.drop(['iPhone 7','iPhone 8','iPhone X'], axis='index', inplace = False)
```
