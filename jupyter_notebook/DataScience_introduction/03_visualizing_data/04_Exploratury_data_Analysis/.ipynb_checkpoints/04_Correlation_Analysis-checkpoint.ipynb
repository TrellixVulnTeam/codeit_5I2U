# 04_상관관계분석

young_survey.csv 파일에 대한 설명

147개의 항목을 조사하얐고 997명이 참여하였다(147 columns,997 rows)

columns의 소개

    -0~18 : 음악 취향
    -19~30 : 영화 취향
    -31~62 : 취미/관심사
    -63~72 : 공포증
    -73~75 : 건강 습관
    -76~132 : 성격, 인생관 등
    -133~139 : 소비 습관
    -140~146 : 기본 정보

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/young_survey.csv')
df.head()
```

음악과 관련된 컬럼부터 한번 봐보자

```python
music = df.iloc[:,:19]
music.head()
```

```python
music.loc[music['Music'] == 5.0,:].describe()
```

각 음악사이의 연관성을 한번 알아보자

```python
sns.heatmap(music.corr())
```

연관성을 알아보는 함수는 df.corr()이다

```python
df.corr()['Age'].sort_values(ascending=False).tail(60)
```
