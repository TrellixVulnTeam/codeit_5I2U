```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/young_survey.csv')
df.head()
```

994명의 학생들의 관심사에 따라 동아리를 묶어주려고 합니다 하지만 관심사가 너무많아 그동아리를 다 만들어줄수없어 관심사비슷한 사람들끼리 묶어서 동아리를 만들어주려고합니다

관심사의 종류는 32개 이다

```python
interests = df.loc[:,'History':'Pets']
interests.head()
```

관심사들끼리의 상관계수를 알고싶다면?

```python
corr = interests.corr()
corr
```

여기서 역사랑 묶으려먼 어떤관심사를 가진사람들과 묶을수 있을까?

일단 corr 에서 역사 컬럼만 따로 꺼내보자

```python
corr['History']
```

위의 결과를 큰순서대로 정렬해보자

```python
corr['History'].sort_values(ascending=False)
```

위의 결과를 보니 역사와 관심사가 비슷한 과목들 상위 몇개가보인다 그런데 이렇게 하나하나 일일이 하기엔 아주 긔찮기 떄문에 한번에 여러개씩 묶어주는 편리한 방법이있다 그것이바로 seaborn 라이브러리에 있는 클러스터맵 이다

```python
sns.clustermap(corr)
```
