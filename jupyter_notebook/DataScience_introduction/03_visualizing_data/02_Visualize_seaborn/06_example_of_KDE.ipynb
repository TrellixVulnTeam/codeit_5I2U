KDE 다양한 그래프에 쓰일수 있다

```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('data/body.csv', index_col=0)
body_df
```

키에 대한 히스토그램을 사용해보자

```python
body_df.plot(kind='hist',y='Height',bins = 15)
```

seaborn을 이용하여 히스토그램위에 kde 를 얹을수 있다

```python
sns.distplot(body_df['Height'],bins=15)
```

이번엔 박스플롯에 한번 적용해보자

```python
body_df.plot(kind='box', y='Height')
```

```python
sns.violinplot(y=body_df['Height'])
```

박스보다 분포가 고루고루 잘보인다.

키와 몸무개의 연관성도 한번 보자

```python
body_df.plot(kind='scatter',x='Height', y='Weight')
```

특정지점이 유난히 몰려있음 특히 70kg 175cm 이부분이 가장 몰려있다 이걸좀더 멋있게 보는방법이 있다 보자

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
```

등고선과 흡사하다 어케 보는걸까? 일단 키 몸무개 각각의 산정도에대한 ked 를 살펴보자

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Weight'])
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
```

위의 두 kdf를 3차원으로 표시한것이다 와우 선간격이 좁을수록 가파르다는것이고 넓을수록 평평하다는 것이다
